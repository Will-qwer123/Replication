```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# Load cleaned data
cleaned_df = pd.read_csv("../data/cleaned_data_preprocessing.csv",keep_default_na=False)

# Stratification variable
mandate_period_label = cleaned_df["within_mandate_period"]

# Split training and test set
df_train, df_test = train_test_split(
    cleaned_df,
    test_size=0.2,
    random_state=42,
    stratify=mandate_period_label
)

# Save train and test datasets
df_train.to_csv("../data/df_train.csv", index=False)
df_test.to_csv("../data/df_test.csv", index=False)


# Function for saving model datasets
def save_model_data(
    model_name,
    response_col,
    drop_cols,
    train_filter=None,
    test_filter=None
):
    # Select rows
    if train_filter is None:
        train_data = df_train.copy()
    else:
        train_data = df_train.loc[train_filter].copy()

    if test_filter is None:
        test_data = df_test.copy()
    else:
        test_data = df_test.loc[test_filter].copy()

    # Select feature columns
    feature_cols = cleaned_df.columns.drop(drop_cols)

    X_train = train_data[feature_cols]
    X_test = test_data[feature_cols]

    # Encode target variable
    label_encoder = LabelEncoder()

    # Use the same encoder for both train and test labels
    y_train = label_encoder.fit_transform(train_data[response_col].values.ravel())
    y_test = label_encoder.transform(test_data[response_col].values.ravel())

    # Save files
    X_train.to_csv(f"../data/X_train_model_{model_name}.csv", index=False)
    X_test.to_csv(f"../data/X_test_model_{model_name}.csv", index=False)

    pd.DataFrame({"y_train": y_train}).to_csv(
        f"../data/y_train_model_{model_name}.csv", index=False)

    pd.DataFrame({"y_test": y_test}).to_csv(
        f"../data/y_test_model_{model_name}.csv", index=False)

    # Check output
    print(f"Model {model_name} saved.")
    print(f"X_train shape: {X_train.shape}")
    print(f"X_test shape:  {X_test.shape}")
    print(f"y_train shape: {y_train.shape}")
    print(f"y_test shape:  {y_test.shape}")
    print("\n")

# Model 1: Face masks
save_model_data(
    model_name="1",
    response_col="face_mask_behaviour_binary",
    drop_cols=[
        "RecordNo",
        "face_mask_behaviour_scale",
        "protective_behaviour_scale",
        "face_mask_behaviour_binary",
        "protective_behaviour_binary",
        "endtime"
    ]
)


# Model 1a: Before mandates
mandate_starter = "2022-01-01"

train_filter_1a = ((df_train["endtime"] < mandate_starter) & (df_train["within_mandate_period"] == 0))
test_filter_1a = ((df_test["endtime"] < mandate_starter) & (df_test["within_mandate_period"] == 0))

save_model_data(
    model_name="1a",
    response_col="face_mask_behaviour_binary",
    drop_cols=[
        "RecordNo",
        "face_mask_behaviour_scale",
        "protective_behaviour_scale",
        "face_mask_behaviour_binary",
        "protective_behaviour_binary",
        "endtime",
        "within_mandate_period"
    ],
    train_filter=train_filter_1a,
    test_filter=test_filter_1a
)


# Model 1b: After mandates
train_filter_1b = df_train["within_mandate_period"] == 1
test_filter_1b = df_test["within_mandate_period"] == 1

save_model_data(
    model_name="1b",
    response_col="face_mask_behaviour_binary",
    drop_cols=[
        "RecordNo",
        "face_mask_behaviour_scale",
        "protective_behaviour_scale",
        "face_mask_behaviour_binary",
        "protective_behaviour_binary",
        "endtime",
        "within_mandate_period"
    ],
    train_filter=train_filter_1b,
    test_filter=test_filter_1b
)


# Model 2: Protective behaviour
save_model_data(
    model_name="2",
    response_col="protective_behaviour_binary",
    drop_cols=[
        "RecordNo",
        "face_mask_behaviour_scale",
        "protective_behaviour_scale",
        "face_mask_behaviour_binary",
        "protective_behaviour_binary",
        "protective_behaviour_nomask_scale",
        "endtime"
    ]
)


# Model 2a: Before mandates
train_filter_2a = ((df_train["endtime"] < mandate_starter) & (df_train["within_mandate_period"] == 0))

test_filter_2a = ((df_test["endtime"] < mandate_starter) & (df_test["within_mandate_period"] == 0))

save_model_data(
    model_name="2a",
    response_col="protective_behaviour_binary",
    drop_cols=[
        "RecordNo",
        "face_mask_behaviour_scale",
        "protective_behaviour_scale",
        "face_mask_behaviour_binary",
        "protective_behaviour_binary",
        "protective_behaviour_nomask_scale",
        "endtime",
        "within_mandate_period"
    ],
    train_filter=train_filter_2a,
    test_filter=test_filter_2a
)


# Model 2b: After mandates
train_filter_2b = df_train["within_mandate_period"] == 1
test_filter_2b = df_test["within_mandate_period"] == 1

save_model_data(
    model_name="2b",
    response_col="protective_behaviour_binary",
    drop_cols=[
        "RecordNo",
        "face_mask_behaviour_scale",
        "protective_behaviour_scale",
        "face_mask_behaviour_binary",
        "protective_behaviour_binary",
        "protective_behaviour_nomask_scale",
        "endtime",
        "within_mandate_period"
    ],
    train_filter=train_filter_2b,
    test_filter=test_filter_2b
)
```
