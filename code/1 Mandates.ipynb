```python
import pandas as pd

# Load data
df = pd.read_csv("../raw/OxCGRT_AUS_latest.csv")

# Keep only the columns needed for this step
col_subsets = ["RegionName", "RegionCode", "Date", "H6M_Facial Coverings"]
df = df.loc[:, col_subsets].copy()

# Change the Date column to datetime
df.index = pd.to_datetime(df["Date"], format="%Y%m%d")

# Calculate rolling averages
rolling_days = 14
df_rolling = (
    df.loc[:, ["RegionName", "H6M_Facial Coverings"]]
      .groupby("RegionName")
      .rolling(window=rolling_days)
      .mean()
)

# Find the first date when mask mandates become consistent
mandate_limit = 3
df_mandates = (df_rolling.loc[df_rolling["H6M_Facial Coverings"] >= mandate_limit].groupby("RegionName").head(1))

df_mandates.to_csv("../data/mandate_start_dates.csv")
```
