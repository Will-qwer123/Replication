```python
import pandas as pd

# Load cleaned data
cleaned_df = pd.read_csv("../data/cleaned_data.csv", keep_default_na=False)

# Convert endtime to datetime format
cleaned_df["endtime"] = pd.to_datetime(cleaned_df["endtime"], format="%Y-%m-%d")

# Load mandate start dates
mandate_df = pd.read_csv("../data/mandate_start_dates.csv")
mandate_df["Date"] = pd.to_datetime(mandate_df["Date"], format="%Y-%m-%d")

# Rename columns
mandate_df = mandate_df.rename(columns={"RegionName": "state", "Date": "mandate_start_date"})

# Add mandate start date to each record
cleaned_df = cleaned_df.merge(mandate_df[["state", "mandate_start_date"]], on="state", how="left")

# Create mandate period indicator
cleaned_df["within_mandate_period"] = (cleaned_df["endtime"] >= cleaned_df["mandate_start_date"]).astype(int)

# Drop mandate_start_date
cleaned_df = cleaned_df.drop(columns=["mandate_start_date"])

# Select categorical variables
convert_into_dummy_cols = [
    "state",
    "gender",
    "i9_health",
    "employment_status",
    "i11_health",
    "WCRex1",
    "WCRex2",
    "PHQ4_1",
    "PHQ4_2",
    "PHQ4_3",
    "PHQ4_4",
    "d1_comorbidities"
]
# Convert categorical variables into dummy variables
cleaned_df = pd.get_dummies(cleaned_df, columns=convert_into_dummy_cols,drop_first=True)

# Save
cleaned_df.to_csv("../data/cleaned_data_preprocessing.csv", index=False)
```
